```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Task 1: Import the cars-1.csv file in the Jupyter notebook
df = pd.read_csv('cars-1.csv')
```

```python
#2. Generate a line plot graph for the column ‘model’ and ‘hp’: 
#a. Map the ‘model’ column on the x-axis 
#b. Map the ‘hp’ column on the y-axis 
#c. Provide the x-axis label as Models of the cars 
#d. Provide the y-axis label as Horse-Power of Cars 
#e. Set the title as Model Names vs HorsePower

plt.figure(figsize=(10, 6))
plt.plot(df['model'], df['hp'])
plt.xlabel('Models of the cars')
plt.ylabel('Horse-Power of Cars')
plt.title('Model Names vs HorsePower')
plt.xticks(rotation=90)
plt.show()
```

```python
#3. Generate a bar plot graph for the columns ‘carb’ and ‘gear’:
#a. Map the ‘carb’ onto the x-axis
#b. Map the ‘gear’ onto the y-axis 
#c. Provide the x-axis label as Number of carburetors 
#d. Provide the y-axis label as Number of forward gears 
#e. Set the title as carb vs gear

plt.figure(figsize=(8, 6))
plt.bar(df['carb'], df['gear'])
plt.xlabel('Number of carburetors')
plt.ylabel('Number of forward gears')
plt.title('carb vs gear')
plt.show()
```

```python
#4. Plot a histogram for the column ‘wt’: 
#a. Map the ‘wt’ onto the x-axis 
#b. Provide the x-axis label as ‘weight of the cars’ 
#c. Provide the y-axis label as ‘Count’ 
#d. Set the number of bins as 30 
#e. Set the title as ‘Histogram for the weight v

plt.figure(figsize=(8, 6))
plt.hist(df['wt'], bins=30)
plt.xlabel('Weight of the cars')
plt.ylabel('Count')
plt.title('Histogram for the weight')
plt.show()
```

```python
#5. Plot a pie chart for columns: ‘cyl’ and ‘model’ form the mtcars.csv data frame.

plt.figure(figsize=(8, 6)) 
plt.pie(df['cyl'], labels=df['model'], autopct='%1.1f%%') 
plt.title('Pie Chart for cyl and model') 
plt.axis('equal') 
plt.show() 
```

```python
#6. Plot the area chart for the columns ‘am’ and ‘carb’: 
#a. Set the ‘am’ on the x-axis 
#b. Set the ‘carb’ on the y-axis 
#c. Provide the x-axis label as Transmission 
#d. Provide the y-axis label as Number of carburetors
#e. Provide title as Transmission vs. carburetors.

plt.figure(figsize=(10, 6))
plt.plot(df['am'], df['carb'], 'o-', color="blue")
plt.xlabel('Transmission')
plt.ylabel('Number of carburetors')
plt.title('Transmission vs. Carburetors')
plt.show()
```
